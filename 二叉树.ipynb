### 二叉树及其性质

满足以下两条性质:

+ 本身是有序树
+ 树中包含的各节点的度不能超过 2, 即 0、1 或者 2.

```python
from IPython.display import Image
Image(filename='./figures/two_tree.png',width=600, height=400)
```

图 1 (a) 是二叉树, 图 1 (b) 不是二叉树.

性质:

+ 二叉树中，第 $n$ 层最多有 $2n-1$ 个结点。
+ 如果二叉树的深度为 $d$, 那么此二叉树最多有 $2d-1$ 个结点。
+ 二叉树中, 终端结点数（叶子结点数）为 $n_0$, 度为 2 的结点数为 $n_2$, 则 $n_0=n_2+1$.

满二插叉树

如果二叉树中**除去最后一层节点**为满二叉树，且最后一层的结点依次从**左到右分布**，则此二叉树被称为完全二叉树。

```python
from IPython.display import Image
Image(filename='./figures/do_twotree.png',width=600, height=400)
```

如图 3a) 所示是一棵完全二叉树，图 3b) 由于最后一层的节点没有按照从左向右分布, 因此只能算作是普通的二叉树.

完全二叉树除了具有普通二叉树的性质, 它自身也具有一些独特的性质, 比如说, $n$ 个结点的完全二叉树的深度为 $[log2n]+1$.
