### 输出某个路径及其子目录下的素有文件路径

```python
import os
import numpy as np
```

```python
def show_dir(filepath):
    for i in os.listdir(filepath):
        path = (os.path.join(filepath, i))
        print(path)
        if os.path.isdir(path): # isdir() 判断是否为目录
            show_dir(path) # 如果是目录使用递归的方法
```
